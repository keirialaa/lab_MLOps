```python
# Generate all categories

if input("Generate new inventory? (y/n): ").lower() == "y":
    gen = fn.generate()
    results = {}
    start_item_id = 1

    for name, cfg in fn.category_config.items():
        df = gen.example_product(
            category=name,
            amount=cfg["amount"],
            loan_percentage=cfg["loan_percentage"],
            oddity_percentage=cfg["oddity_percentage"],
            start_item_id=start_item_id,
            category_id=fn.category_id[name],
            dataframe=True
        )
        results[name] = df
        start_item_id += len(df)  # advance by however many items were generated

fn.generate().bulk_insert(
    dict_of_df=results,
    file_name_and_path=os.path.join(fn.PWD, "database", "pawnshop.db"),
    table_name="inventory",
    if_exists="replace"
)
```
